```python
import  pandas as pd
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
df=pd.read_csv("../data/B_customer_churn_cleaned.csv")
x=df[['complaints_count', 'monthly_spend']]
y=df['is_churned']
clf= DecisionTreeClassifier( max_depth=2,random_state=42)
clf.fit(x, y)

plt.figure(figsize=(12, 8),dpi=150)
plot_tree(
    clf,
    feature_names=x.columns,
    class_names=['Stay', 'Chur'],
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree for Customer Churn Prediction")
plt.show()
df['high_complaints'] = df['complaints_count'] >= 3
churn_rate = df.groupby('high_complaints')['is_churned'].mean() * 100
print("Гомдлын тоо 3-аас бага хэрэглэгчдийн цуцлах магадлал: {:.1f}%".format(churn_rate[False]))
print("Гомдлын тоо 3 ба түүнээс дээш хэрэглэгчдийн цуцлах магадлал: {:.1f}%".format(churn_rate[True]))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df, 
    x='monthly_spend', 
    y='monthly_usage_gb', 
    hue='is_churned', 
    palette='coolwarm', 
    alpha=0.6
)
plt.title('Monthly Spend vs Monthly Usage GB by Churn Status')
plt.xlabel('Monthly Spend')
plt.ylabel('Monthly Usage (GB)')   
plt.show()
```
