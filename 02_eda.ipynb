```python
import pandas as pd

df = pd.read_csv("../data/B_customer_churn.csv")
print("Анхны датаны хэмжээ:", df.shape)
print(df.isna().sum())
df['monthly_spend'] = df['monthly_spend'].fillna(df['monthly_spend'].median())
df['monthly_usage_gb'] = df['monthly_usage_gb'].fillna(df['monthly_usage_gb'].median())
print("\nЦэвэрлэсний дараах хоосон утга:", df.isna().sum().sum())
```

```python
df["plan_type"] = df["plan_type"].str.strip().str.lower()
print(df["plan_type"].value_counts())
```

```python
print("Давхардсан мөрний тоо:", df.duplicated().sum())

df= df.drop_duplicates()
print(("Цэвэрлэсний дараах" " давхардсан мөрний тоо:", df.duplicated().sum()))
```
